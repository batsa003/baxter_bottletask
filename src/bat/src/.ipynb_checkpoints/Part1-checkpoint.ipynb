```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
import os
%matplotlib inline

# Constants:
vision_x = 0.75
vision_y = -0.2
vision_z = 0.163
table_z = -0.265

def move_tool(x,y,z):
    command = 'python example.py %0.3f %0.3f %0.3f' % (x, y, z)
    print('Executing: ' + command)
    os.system(command)
```

```python
# Step 1. 
move_tool(vision_x, vision_y, vision_z)

# Step 2.Save camera file to latest.png
```

```python
filename = 'latest.png'
im = cv2.imread(filename)

# Step 3. Locate points manually, and circle them.
obj_x = 670
obj_y = 360
cv2.circle(im,(obj_x,obj_y),10,(0,0,255),2)

obj2_x = 845
obj2_y = 450
cv2.circle(im,(obj2_x,obj2_y),10,(0,255,0),2)
plt.imshow(im)
plt.show()
```

```python
def f(xx,yy):
    pix_size = .0023 #Camera calibration (meter/pixels); Pixel size at 1 meter height of camera
    h = 0.433 #Height from table to camera, when at vision place
    x0b = vision_x # x position of camera in baxter's base frame, when at vision place 
    y0b = vision_y # y position of camera in baxter's base frame, when at vision place
    #x_camera_offset = .02 #x camera offset from center of the gripper  
    #y_camera_offset = -.02 #y camera offset from center of the gripper 
    #x_camera_offset = .045 #x camera offset from center of the gripper  
    #y_camera_offset = -0.01
    x_camera_offset = 0.0
    y_camera_offset = 0.0
    # height, width, depth = cv_image.shape #camera frame dimensions
    depth = 3
    height = 800
    width = 1280
    #Position of the object in the baxter's stationary base frame
    xb = (yy - (height/2))*pix_size*h + x0b + x_camera_offset 
    yb = (xx - (width/2))*pix_size*h + y0b  + y_camera_offset
    print(xb,yb)
    return (xb,yb)
f(obj_x, obj_y) # ideally,  0.82 -0.32 -0.265
print
f(obj_x, 800-obj_y) # ideally,  0.82 -0.32 -0.265
print
f(1280-obj_x, obj_y) # ideally,  0.82 -0.32 -0.265
print
f(1280-obj_x, 800-obj_y) # ideally,  0.82 -0.32 -0.265

#f(obj2_x, obj2_y) # ideally, 0.68 -0.04 -0.2
print
print
print

f(obj2_x, obj2_y) # ideally, 
print
f(obj2_x, 800-obj2_y) # ideally, 
print
f(1280-obj2_x, obj2_y) # ideally,  
print
f(1280-obj2_x, 800-obj2_y) # ideally,
```
