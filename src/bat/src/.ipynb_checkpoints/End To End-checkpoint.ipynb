```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
import os
%matplotlib inline

# Make sure you change the resolution of the camera to 960 by 600.
#rosrun baxter_tools camera_control.py -o right_hand_camera -r 960x600

#import baxter_interface
#self.gripper = baxter_interface.Gripper("right")
#self.gripper.calibrate()

# Constants:
# Coordinates of the camera at the initial location
vision_x = 0.75 
vision_y = -0.2
vision_z = 0.163
# Height of the table
table_z = -0.265

# Helper function to move the tool to (x,y,z)
def move_tool(x,y,z):
    command = 'python ik_helper.py %0.3f %0.3f %0.3f' % (x, y, z)
    print('Executing: ' + command)
    os.system(command)
```

```python
def pixel_to_world(xx,yy,h=0.428):
    h+= 0.07
    pix_size = .0025 #Camera calibration (meter/pixels); Pixel size at 1 meter height of camera
    x0b = vision_x # x position of camera in baxter's base frame, when at vision place 
    y0b = vision_y # y position of camera in baxter's base frame, when at vision place
    #x_camera_offset = .02 #x camera offset from center of the gripper  
    #y_camera_offset = -.02 #y camera offset from center of the gripper 
    #x_camera_offset = .045 #x camera offset from center of the gripper  
    #y_camera_offset = -0.01
    x_camera_offset = 0.015
    y_camera_offset = -0.045
    # height, width, depth = cv_image.shape #camera frame dimensions
    depth = 3
    #Position of the object in the baxter's stationary base frame
    xb = (yy - (height/2))*pix_size*h + x0b + x_camera_offset 
    yb = (xx - (width/2))*pix_size*h + y0b  + y_camera_offset
    print(xb,yb)
    return (xb,yb)
```

```python
#print pt_cap[0], pt_cap[1]
(xc, yc) = pixel_to_world(pt_cap[0],pt_cap[1])
```
