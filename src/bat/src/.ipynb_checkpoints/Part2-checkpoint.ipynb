```python
def f(xx,yy):
    pix_size = .0023 #Camera calibration (meter/pixels); Pixel size at 1 meter height of camera
    h = .165 #Height from table to camera, when at vision place
    x0b = vision_x # x position of camera in baxter's base frame, when at vision place 
    y0b = vision_y # y position of camera in baxter's base frame, when at vision place
    #x_camera_offset = .02 #x camera offset from center of the gripper  
    #y_camera_offset = -.02 #y camera offset from center of the gripper 
    #x_camera_offset = .045 #x camera offset from center of the gripper  
    x_camera_offset = 0.0
    y_camera_offset = 0.0
    # height, width, depth = cv_image.shape #camera frame dimensions
    depth = 3
    height = 800
    width = 1280
    #Position of the object in the baxter's stationary base frame
    xb = (yy - (height/2))*pix_size*h + x0b + x_camera_offset 
    yb = (xx - (width/2))*pix_size*h + y0b  + y_camera_offset
    print(xb,yb)
    return (xb,yb)
f(obj_x, obj_y) 
print
f(obj_x, 800-obj_y) 
print
f(1280-obj_x, obj_y) 
print
f(1280-obj_x, 800-obj_y) 
```
