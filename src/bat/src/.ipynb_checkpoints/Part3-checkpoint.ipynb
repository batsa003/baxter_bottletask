```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
import os
%matplotlib inline

# Constants:
vision_x = 0.75
vision_y = -0.2
vision_z = 0.163
table_z = -0.265

def move_tool(x,y,z):
    command = 'python example.py %0.3f %0.3f %0.3f' % (x, y, z)
    print('Executing: ' + command)
    os.system(command)
```

```python
# Step 1. 
move_tool(vision_x, vision_y, vision_z)

# Step 2.Save camera file to latest.png
```

```python
filename = 'latest.png'
im = cv2.imread(filename)
height,width,depth = im.shape

# Step 3. Locate points manually, and circle them.
x1 = 590
y1 = 170
# Circle BLUE
cv2.circle(im,(x1,y1),10,(0,0,255),2) # RED

x2 = 510
y2 = 310
# Circle GREEN
cv2.circle(im,(x2,y2),10,(0,255,0),2) # ORANGE

x3 = 270
y3 = 263
#Circle RED
cv2.circle(im,(x3,y3),10,(255,0,0),2) # VIOLET
plt.imshow(im)
plt.show()
```

```python
def f(xx,yy):
    pix_size = .00287 #Camera calibration (meter/pixels); Pixel size at 1 meter height of camera
    h = 0.433 #Height from table to camera, when at vision place
    x0b = vision_x # x position of camera in baxter's base frame, when at vision place 
    y0b = vision_y # y position of camera in baxter's base frame, when at vision place
    #x_camera_offset = .02 #x camera offset from center of the gripper  
    #y_camera_offset = -.02 #y camera offset from center of the gripper 
    #x_camera_offset = .045 #x camera offset from center of the gripper  
    #y_camera_offset = -0.01
    x_camera_offset = 0.007
    y_camera_offset = -0.04
    # height, width, depth = cv_image.shape #camera frame dimensions
    depth = 3
    #Position of the object in the baxter's stationary base frame
    xb = (yy - (height/2))*pix_size*h + x0b + x_camera_offset 
    yb = (xx - (width/2))*pix_size*h + y0b  + y_camera_offset
    print(xb,yb)
    return (xb,yb)
f(x1, y1)
print
f(x1, height-y1) 
print
f(width-x1, y1) 
print
f(width-x1, height-y1)
print
print
print

f(x2, y2)
print
f(x2, height-y2) 
print
f(width-x2, y2) 
print
f(width-x2, height-y2)
print
print
print

f(x3, y3)
print
f(x3, height-y3) 
print
f(width-x3, y3) 
print
f(width-x3, height-y3)
print
print
print


#move_tool(0.6, -0.1,-0.1) # RED
#move_tool(0.78, -0.2, -0.26) # ORANGE
#move_tool(0.7, -0.5, -0.26) # VIOLET
```
